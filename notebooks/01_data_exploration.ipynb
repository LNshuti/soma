```python
import pandas as pd 
import numpy as np 
import duckdb
```

```python
books_db = duckdb.connect('../data/soma.duckdb')
books_db
```

```python
books = books_db.execute("""
                SELECT *
                --book_id, title, genre, price 
                FROM dim_books 
                ORDER BY title 
                LIMIT 100
            """).df()
books
```

```python
stg_books_df = books_db.execute(
    """ 
        SELECT         
            *
        FROM main.stg_books b 
    """
    ).df()
stg_books_df
```

```python
# Load sales data as user-item interactions
sales_df = books_db.execute("""
    SELECT 
        customer_type as user_id,
        book_id,
        SUM(quantity) as rating,
        COUNT(*) as interaction_count
    FROM fact_sales 
    WHERE sale_date >= CURRENT_DATE - INTERVAL '365 days'
    GROUP BY customer_type, book_id
    HAVING SUM(quantity) > 0
""").df()

sales_df            
```

```python
# Load book features
features_query = books_db.execute("""
    SELECT 
        book_id,
        genre,
        price_category,
        length_category,
        recency_category,
        publisher_type,
        avg_selling_price,
        total_quantity_sold,
        velocity_score
    FROM book_features
    WHERE total_transactions > 0
""").df()

features_query
```
